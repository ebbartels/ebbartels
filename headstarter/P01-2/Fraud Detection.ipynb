```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle

import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import GradientBoostingClassifier
```

```python
train_df.columns
```

```python
X_cols = ["category", "amt", "unix_time"]
y_col = ["is_fraud"]
```

```python
print(train_df["merchant"].value_counts())
print(train_df["category"].value_counts())
print(train_df["job"].value_counts())
```

```python
X_train = train_df[X_cols]
X_train = pd.get_dummies(X_train, columns=["category"])
y_train = train_df[y_col]
y_train_flat = train_df[y_col].values.flatten()  # Also converts DataFrame to 1D array

X_test = test_df[X_cols]
X_test = pd.get_dummies(X_test, columns=["category"])
y_test = test_df[y_col]
y_test_flat = test_df[y_col].values.flatten()
```

```python
X_train
```

```python
def evaluate_and_save_model(model, X_train, X_test, y_train, y_test, filename):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    print(f"{model.__class__.__name__} Accuracy: {accuracy:.4f}")
    print(f"\nClassification Report:\n{classification_report(y_test, y_pred)}")
    print("------------")
    
    with open(filename, "wb") as f:
        pickle.dump(model,f)
    
    print(f"Model saved as {filename}")
```

```python
xbg_model = xgb.XGBClassifier(random_state=42)
evaluate_and_save_model(xbg_model, X_train, X_test, y_train, y_test, "xgb_model.pkl")

dtc_model = DecisionTreeClassifier(random_state = 42)
evaluate_and_save_model(dtc_model, X_train, X_test, y_train, y_test, "dtc_model.pkl")

rfc_model = RandomForestClassifier(random_state = 42)
evaluate_and_save_model(rfc_model, X_train, X_test, y_train_flat, y_test_flat, "rfc_model.pkl")

#svm_model = SVC(random_state = 42)
#evaluate_and_save_model(svm_model, X_train, X_test, y_train_flat, y_test_flat, "svm_model.pkl")
```

```python
lr_model = LogisticRegression(random_state = 42)
evaluate_and_save_model(lr_model, X_train, X_test, y_train, y_test, "lr_model.pkl")
```

```python
gb_modle = GradientBoostingClassifier(random_state = 42)
evaluate_and_save_model(gb_modle, X_train, X_test, y_train, y_test, "gb_modle.pkl")
```

```python
smote = SMOTE(random_state = 42)
```

```python
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

```python
evaluate_and_save_model(xbg_model, X_resampled, X_test, y_resampled, y_test, "xgboost-SMOTE.pkl")
```

```python
voting_clf = VotingClassifier(
    estimators = [("xgboost", xgb.XGBClassifier(random_state = 42)), ("rf", RandomForestClassifier(random_state= 42)), ("gbG", GradientBoostingClassifier(random_state = 42))],
    voting = "hard"
)
```

```python
evaluate_and_save_model(voting_clf, X_resampled, X_test, y_resampled, y_test, "voting_clf.pkl")
```
